```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc = {'figure.figsize':(12,12)})
```

```python
df=pd.read_csv('Excel/improved-water-sources-egypt-mics-2013-14-sheet1.csv')
df
```

```python
df = df.dropna(axis=0)
df.drop(index=0, inplace=True, axis=0)
df.drop(columns=['Total'], inplace=True, axis=1)
df = df.rename(columns={'Unnamed: 0': 'Category'})
df
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.barplot(y='Category',x='Piped into dwelling',data=df)
```

```python
sns.barplot(y='Category',x='Piped into compound, yard or plot',data=df)
```

```python
sns.barplot(y='Category',x='Piped to neighbour',data=df)
```

```python
df.sort_values("Percentage using improved sources of drinking water", axis = 0, ascending = False,
                 inplace = True, na_position ='last')
df[['Category', 'Percentage using improved sources of drinking water'
]]
```

```python
sns.barplot(y='Category',x='Percentage using improved sources of drinking water',data=df)
```

```python
fig, ax = plt.subplots(figsize=(13,8))
df.drop(['Piped into dwelling', 'Piped into compound, yard or plot', 'Piped to neighbour', 'Percentage using improved sources of drinking water', 'Number of household members'], axis=1).groupby('Category').sum().sort_values(by='Category', ascending=False).plot(ax=ax, kind='bar', lw = 1.5)
```

```python
df.corr()
sns.heatmap(df.corr(), cmap='Blues', annot=True, linewidths=1,
  linecolor='black',) 
```

- The tube well is more common than other *other improved* sources
